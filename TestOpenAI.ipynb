```python
## Chat Completetion API and Completion API

## Completion API

response = openai.Completion.create(
    model='text-curie-001',
    prompt="Explain Machine Learning to a 5 years old"
)

response
```

```python
## Chat Completetion API and Completion API

## Completion API

response = openai.Completion.create(
    model='text-curie-001',
    prompt="Act as an AI assistant and Explain Machine Learning to a 5 years old",
    max_tokens=25,
    temperature=0.6,
    n=5
)

response
```

```python
response = openai.Completion.create(
    model='text-davinci-003',
    prompt="""Act as an AI model and detect positive and negative tweets:
    This person is bad and he is biased towards negative sentiments""",
    max_tokens=15,
    temperature=1,
    n=1
)

print(response)
```

```python
#Chat Completion API
openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=[{'role':'system','content':"Act as an AI assistant"},
              {'role':'user', 'content':"Hello How are You"}]
)
```

```python
#Chat Completion API
openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=[{'role':'system','content':"Act as an AI assistant"},
              {'role':'user', 'content':"Hello How are You"},
              {'role':'user', 'content':"I turned 23 on March 9th this year 2023"},
              {'role':'user', 'content':"can you tell me what is my date of birth, in the format of 'DD/MM/YYYY'?"}]
)
```
